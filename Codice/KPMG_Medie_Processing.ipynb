```python
with open("/content/drive/MyDrive/KPMG/RegressionDataset.xlsx", 'rb') as fp:
    df = pd.read_excel(fp)
```

```python
df
```

```python
lista_atenei=df['Ateneo'].unique()
```

```python
lista_fuorisede_ateneo=[]
dizionario_lista_differenze={}

for ateneo in lista_atenei[:]:
    df_mask=df['Ateneo']==ateneo
    filtered_df = df[df_mask]
    lista_fuorisede_ateneo=list(filtered_df['Fuorisede'])
    temp={ateneo: lista_fuorisede_ateneo}
    dizionario_lista_differenze.update(temp)
```

```python
lista_differenze=[]
dizionario={}

for ateneo in lista_atenei[:]:
  for i in range(0,len(dizionario_lista_differenze[ateneo])-1):
    risultato=dizionario_lista_differenze[ateneo][i+1]-dizionario_lista_differenze[ateneo][i]
    lista_differenze.append(risultato)
  #temp={ateneo: lista_differenze}
  mean = sum(lista_differenze)/len(lista_differenze)
  lista_differenze=[]
  temp={ateneo: mean}
  dizionario.update(temp)

```

```python
df_fuorisede = pd.DataFrame(list(dizionario.items()),columns = ['Ateneo','Fuorisede']) 
```
