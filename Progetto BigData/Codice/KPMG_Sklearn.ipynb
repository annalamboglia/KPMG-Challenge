```python
X=df.drop(columns=['Immatricolati'])
Y=df["Immatricolati"]
```

```python
#Split the Training 75% and 25% Testing
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
#Split the Training 75% and Validation 25%
from sklearn.model_selection import train_test_split

X_train, X_validation, Y_train, Y_validation = train_test_split(X_train, Y_train, test_size=0.25, random_state=0)
```

```python
#Scale the data (Feature Scaling Z Score) z = (x - u) / s
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_validation=sc.transform(X_validation)
X_test = sc.transform(X_test)
```

```python
X_train = np.array(X_train)
X_validation=np.array(X_validation)
X_test = np.array(X_test)

Y_train = np.array(Y_train)
Y_validation=np.array(Y_validation)
Y_test = np.array(Y_test)
```

#RandomForest Regressor

```python
import time
start = time.time()

from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=1000,max_depth=3, random_state=0) 

regressor.fit(X_train,Y_train)

end = time.time()
print(format(end-start))
```

```python
# test the output by changing values, like 3750
y_pred = regressor.predict(X_validation)
  
# print the predicted price
print("Predicted Immatricolazioni:")
print(y_pred) 
```

```python
from sklearn.metrics import mean_absolute_error

y_pred = regressor.predict(X_validation)
print(mean_absolute_error(Y_validation, y_pred))

from sklearn.metrics import r2_score

#calculate R-squared of regression model
print("\tR2 score:", r2_score(Y_validation, y_pred))
```

```python
from sklearn.metrics import mean_absolute_error

y_pred_test = regressor.predict(X_test)
print(mean_absolute_error(Y_test, y_pred_test))

from sklearn.metrics import r2_score

#calculate R-squared of regression model
print("\tR2 score:", r2_score(Y_test, y_pred_test))
```

```python
X=df.drop(columns=['Immatricolati'])
Y=df["Immatricolati"]
```

```python
#Split the training e validation 75% and 25% testing
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=0)
```

```python
#Scale the data (Feature Scaling Z Score) z = (x - u) / s
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)

Y_train = np.array(Y_train)
Y_test = np.array(Y_test)
```

```python
import time
start = time.time()

from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=1000,max_depth=3, random_state=0) 

regressor.fit(X_train,Y_train)

end = time.time()
print(format(end-start))
```

```python
import time
start = time.time()

from sklearn.ensemble import ExtraTreesRegressor
regressor = ExtraTreesRegressor(n_estimators=400,random_state=0)

regressor.fit(X_train,Y_train)

end = time.time()
print(format(end-start))
```

```python
# test the output by changing values, like 3750
y_pred_test = regressor.predict(X_test)
  
# print the predicted price
print("Predicted Immatricolazioni:")
print(y_pred_test) 
```

```python
from sklearn.metrics import mean_absolute_error

y_pred_test = regressor.predict(X_test)
print(mean_absolute_error(Y_test, y_pred_test))
```

```python
from sklearn.metrics import r2_score

#calculate R-squared of regression model
print("\tR2 score:", r2_score(Y_test, y_pred_test))
```
